```python

theta_marker = ["{}:00".format(i) for i in range(24)]
timeStamp = [
'子正', '丑初', '丑正', '寅初', '寅正', 
'卯初', '卯正', '辰初', '辰正', '巳初', '巳正',  
'午初', '午正', '未初', '未正', '申初', '申正',  
'酉初', '酉正', '戌初', '戌正', '亥初', '亥正', '子初'
]
def sector(name, location, radius,  fillcolor):
    r = [0] * 24
    r[location % 24] = radius
    r[(location + 1) % 24] = radius
    obj = go.Scatterpolar(
      name = name,
      r = r,
      theta = theta_marker,
      fill = "toself",
      fillcolor = fillcolor
    )
    return obj




trace = [sector(i, j, 5) for (i,j) in zip(timeStamp, range(24))]
trace += [go.Scatterpolar(
              name = "test",
              r = [3,3, *[0] * 21, 3],
              theta = theta_marker,
              fill = "toself",
              line = dict(color='rgb(0,0,0)'),
              fillcolor = 'rgb(0,0,0)',
              opacity=0.8
            )]
layout = go.Layout(
    polar = dict(
      radialaxis = dict(
        visible = False
      ),
      angularaxis = dict(
        direction = "clockwise",
        visible = True,
        linewidth = 3
      )
    )
)


plotly.offline.iplot({
    "data": trace,
    "layout": layout
})
```
